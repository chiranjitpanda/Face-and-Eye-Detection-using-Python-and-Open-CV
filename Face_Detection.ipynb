# Face Detection

```python
face_cascade = cv2.CascadeClassifier(r'C:\Users\HP\haarcascades\haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
    
    face_image = img.copy()
    face_rect = face_cascade.detectMultiScale(face_image)
    
    for(x,y,w,h) in face_rect:
        cv2.rectangle(face_image,(x,y),(x+w,y+h),(255,0,0),10)
    
    return face_image
```

```python
img1 = detect_face(nadia)
```

```python
plt.imshow(img1,cmap='gray')
```

```python
img2= detect_face(denis)
img3 = detect_face(solvay)
```

```python
plt.imshow(img2,cmap='gray')
```

```python
plt.imshow(img3,cmap='gray')
```

```python
def detect_face2(img):
    
    face_image = img.copy()
    face_rect = face_cascade.detectMultiScale(face_image,scaleFactor=1.2,minNeighbors=5)
    
    for(x,y,w,h) in face_rect:
        cv2.rectangle(face_image,(x,y),(x+w,y+h),(255,0,0),4)
    
    return face_image
```

```python
img4 = detect_face2(solvay)
```

```python
plt.imshow(img4,cmap='gray')
```

# Real Time Face Detection

```python
cap = cv2.VideoCapture(0)

while True:
    ret,frame = cap.read()
    
    frame = detect_face(frame)
    cv2.imshow("Camera",frame)
    if cv2.waitKey(1) == 13:
        
        break
        
cap.release()
cv2.destroyAllWindows()
```

# Eye Detection

```python
eye_cascade = cv2.CascadeClassifier(r'C:\Users\HP\haarcascades\haarcascade_eye.xml')
```

```python
def detect_eye(img):
    
    face_image = img.copy()
    face_rect = eye_cascade.detectMultiScale(face_image)
    
    for(x,y,w,h) in face_rect:
        cv2.rectangle(face_image,(x,y),(x+w,y+h),(255,0,0),6)
    
    return face_image
```

```python
img6 = detect_eye(nadia)
```

```python
plt.imshow(img6,cmap='gray')
```

```python
cap = cv2.VideoCapture(0)

while True:
    ret,frame = cap.read()
    
    frame = detect_eye(frame)
    cv2.imshow("Camera",frame)
    if cv2.waitKey(1) == 13:
        
        break
        
cap.release()
cv2.destroyAllWindows()
```
